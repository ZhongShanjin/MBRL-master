```python
def calarea(j,w,h): #计算边框宽度和高度和面积
    bw=j[2]-j[0]
    bh=j[3]-j[1]
#     xc=(j[0]+bw/2.)/w
#     yc=(j[1]+bh/2.)/h
    xc=(j[0]+bw/2.)
    yc=(j[1]+bh/2.)
    if bw==0:
        bw=1
    if bh==0:
        bh=1
#     area=(bw*bh)/(w*h)
    area=(bw*bh)
    return xc,yc,area

def calloc(boxes1,boxes2,w,h): #计算边框宽度和高度和面积
    lx=np.maximum(boxes1[0],boxes2[0])
    ly=np.maximum(boxes1[1],boxes2[1])
    rx=np.minimum(boxes1[2],boxes2[2])
    ry=np.minimum(boxes1[3],boxes2[3])
    iouw=rx-lx
    iouh=ry-ly
    if iouw==0:
        iouw=1
    if iouh==0:
        iouh=1
    if iouh<0:
        return 0
    if iouw<0:
        return 0
    inter = (iouw*iouh)
    _,__,inner=calarea(boxes1,w,h)
    _,__,outer=calarea(boxes2,w,h)
    if inner>outer:
        ioun=inter/outer
        return ioun
    else:
        ioun=inter/inner
        return -ioun
```

```python
box_topk=10000
rel_topk=10000
# 绘制边界框
def draw_single_box(pic, box, color='blue', draw_info=None):
    draw = ImageDraw.Draw(pic)
    x1,y1,x2,y2 = int(box[0]), int(box[1]), int(box[2]), int(box[3])
    draw.rectangle(((x1, y1), (x2, y2)), outline=color)
    if draw_info:
        draw.rectangle(((x1, y1), (x1+50, y1+10)), fill=color)
        info = draw_info
        draw.text((x1, y1), info)
        
def draw_image(img_path, data, objname, subname):
    # 路径合并
    img_path = os.path.join(img_path)
    pic = Image.open(img_path)
    boxes = np.array(data['boxes'][:box_topk])
    labels = data['labels']
    labels = [str(i)+"-"+vocab['idx_to_label'][str(int(x))] for i, x in enumerate(labels)]
    
    num_obj = boxes.shape[0]
    for i in range(num_obj):
        info = labels[i]
        if objname in info or subname in info:
            draw_single_box(pic, boxes[i], draw_info=info)
    
    all_rel_pairs = data['relations'][:,:2]
    all_rel_labels = data['relations'][:,2]
    
    rel_labels = []
    for i in range(len(all_rel_pairs)):
        if all_rel_pairs[i][0] < box_topk and all_rel_pairs[i][1] < box_topk:
            info=labels[all_rel_pairs[i][0]]+labels[all_rel_pairs[i][1]]
            if objname in info and subname in info:
                label = labels[all_rel_pairs[i][0]] + ', ' + str(int(all_rel_labels[i])) + vocab["idx_to_predicate"][str(int(all_rel_labels[i]))] + ', ' + labels[all_rel_pairs[i][1]]
                rel_labels.append(label)

    rel_labels = rel_labels[:rel_topk]
    for a in list(rel_labels):
        print(a)
    display(pic)
    return None
# 获取图片名
# for i in tqdm(writedics[1:3]):
#     for obj,sub,rel in i['relations']:
#         objname=vocab['idx_to_label'][str(i['labels'][obj])]
#         subname=vocab['idx_to_label'][str(i['labels'][sub])]
#         infolocate=calloc(i['boxes'][obj],i['boxes'][sub],w,h)
# #         if objname=='man' and subname=='man' and infolocate==1:
#         draw_image(i['img_path'], i,objname,subname)
```

```python
def contain(boxes1,boxes2,boxes3,w,h): #obj和sub和UOI是否包含第三者obj或sub
    lx=np.minimum(boxes1[0],boxes2[0])
    ly=np.minimum(boxes1[1],boxes2[1])
    rx=np.maximum(boxes1[2],boxes2[2])
    ry=np.maximum(boxes1[3],boxes2[3])
    xc=(boxes3[0]+boxes3[2])/2.0
    yc=(boxes3[1]+boxes3[3])/2.0
    if (xc>=lx)&(xc<=rx)&(yc>=ly)&(yc<=ry):
        return 1
    return 0
```

```python
def showimg(objname, subname, thrdimname): #显示3D图片
    thrdim=pd.read_csv(thrdimname + '.csv')
    plt.figure(figsize=(20,20))
    ax1 = plt.axes(projection='3d')
    ax1.set_xlabel('X axis')
    ax1.set_ylabel('Y axis')
    ax1.set_zlabel('Z axis')
    cm = plt.cm.get_cmap('jet')
    whobj=(thrdim['objname']==objname) & (thrdim['subname']==subname)
    ax=ax1.scatter(thrdim[whobj]['osxc'],thrdim[whobj]['osyc'],thrdim[whobj]['locate'],c=thrdim[whobj]['rel'],cmap=cm)
    plt.colorbar(ax)
    plt.show()
```

```python
def showcount(thrdimname, percent): #显示关系数量
    thrdim=pd.read_csv(thrdimname + '.csv')
    df_thrdim=thrdim['relname'].value_counts(normalize=percent)
    df_thrdim=pd.DataFrame(df_thrdim)
    df_thrdim.to_csv(thrdimname + 's' + '.csv')
    fig,ax = plt.subplots(figsize=(55,25), dpi=80)
    sns.countplot(x='relname', data=thrdim, order = thrdim['relname'].value_counts().index)
    for p in ax.patches:
        ax.annotate(p.get_height(), (p.get_x(), p.get_height()+50), color='black', size=15)
#     plt.savefig(thrdimname + '.png', dpi=80)
    plt.show()
```

```python
def savethrdim(thrdimname, pk_dir):  #保存三元组
    dics = pickle.load(open(pk_dir, "rb"))
    thrdim=[]
    for i in tqdm(dics):
        w=i['width']
        h=i['height']
        for obj,sub,rel in i['relations']:
            objxc,objyc,objarea=calarea(i['boxes'][obj],w,h)
            subxc,subyc,subarea=calarea(i['boxes'][sub],w,h)
            objname, subname, relname = thename(obj, sub, rel, i)
            osxc=subxc-objxc #sub相对于obj的位置
            osyc=objyc-subyc #sub相对于obj的位置
            locate=calloc(i['boxes'][obj],i['boxes'][sub],w,h)
            thrdim.append([osxc,osyc,objname,subname,relname,rel,locate])
    thrdim=pd.DataFrame(columns=['osxc','osyc','objname','subname','relname','rel','locate'],data=thrdim)
    thrdim.to_csv(thrdimname+'.csv')
    print(len(thrdim))
```

```python
def readpdc(pk_dir, thrdimname): #读取字典，计算关系频率
    dics = pickle.load(open(pk_dir, "rb"))
    writedics = copy.deepcopy(dics)
    thrdim = pd.read_csv(thrdimname + '.csv')
    df_thrdim = thrdim['relname'].value_counts(normalize=True)
    return writedics, df_thrdim
```

```python
def thename(obj,sub,rel,i): #获取物体名和类名
    objname=vocab['idx_to_label'][str(i['labels'][obj])]
    subname=vocab['idx_to_label'][str(i['labels'][sub])]
    relname=vocab['idx_to_predicate'][str(int(rel))]
    return objname, subname, relname
```

```python
def minirel(minumrel, rel,df_thrdim): #获取尾类
    therel=vocab["idx_to_predicate"][str(int(rel))]
    if (minumrel==99):
        return rel
    thisrel=vocab["idx_to_predicate"][str(int(minumrel))]
    if df_thrdim[thisrel]>df_thrdim[therel]:
        minumrel=rel
    return minumrel
```

```python
def thrdim_iol(pk_dir, thrdimname, writedicsname): #添加新标签，针对overlap
    writedics, df_thrdim = readpdc(pk_dir, thrdimname)
    for inx, i in tqdm(enumerate(writedics[:])): #读取字典每一个图片数据
        forerel=-1
        lastrel=0
        while(lastrel>forerel):
            i=writedics[inx]
            w=i['width']
            h=i['height']
            locrel=[] #先保存真实标注
            relations=i['relations']
            forerel=lastrel
            for j_inx, j in enumerate(relations): #标注关系对的位置关系
                objinx,subinx,rel = j
                obj=i['labels'][objinx]
                sub=i['labels'][subinx]
                locate=calloc(i['boxes'][objinx],i['boxes'][subinx],w,h)
                locrel.append([objinx,subinx,rel,obj,sub,locate])
            locrel=np.array(locrel)
            for k_inx, k in enumerate(i['labels']): #枚举关系对
                for l_inx, l in  enumerate(i['labels']):
                    if k_inx==l_inx:
                        continue
                    objinx=k_inx
                    subinx=l_inx
                    obj=k
                    sub=l
                    locate=calloc(i['boxes'][objinx],i['boxes'][subinx],w,h) #物体对位置关系
                    reapt=np.where( (locrel[:,3]==obj) & (locrel[:,4]==sub) & (locrel[:,5]==locate) & \
                               (locrel[:,0]==objinx) & (locrel[:,1]==subinx)) #如果关系对已标注，跳过
                    if len(reapt[0]):
                        continue
                    
                    if (locate==0):
                        reaptobj=np.where( (locrel[:,3]==obj) & (locrel[:,4]==sub) & (locrel[:,5]==locate) & \
                                  (locrel[:,0]==objinx) & (locrel[:,1]!=subinx)) #obj不变，sub变  
                        reaptsub=np.where( (locrel[:,3]==obj) & (locrel[:,4]==sub) & (locrel[:,5]==locate) & \
                                  (locrel[:,0]!=objinx) & (locrel[:,1]==subinx)) #sub不变，obj变
                    elif (locate>0):
                        reaptobj=np.where( (locrel[:,3]==obj) & (locrel[:,4]==sub) & (locrel[:,5]>0) & \
                                      (locrel[:,5]<=locate) & (locrel[:,0]==objinx) & (locrel[:,1]!=subinx)) #obj不变，sub变
                        reaptsub=np.where( (locrel[:,3]==obj) & (locrel[:,4]==sub) & (locrel[:,5]>0) & \
                                      (locrel[:,5]<=locate) & (locrel[:,0]!=objinx) & (locrel[:,1]==subinx)) #sub不变，obj变
                    else:
                        reaptobj=np.where( (locrel[:,3]==obj) & (locrel[:,4]==sub) & (locrel[:,5]<0) & \
                                      (locrel[:,5]>=locate) & (locrel[:,0]==objinx) & (locrel[:,1]!=subinx)) #obj不变，sub变
                        reaptsub=np.where( (locrel[:,3]==obj) & (locrel[:,4]==sub) & (locrel[:,5]<0) & \
                                      (locrel[:,5]>=locate) & (locrel[:,0]!=objinx) & (locrel[:,1]==subinx)) #sub不变，obj变
                    
                    if len(reaptobj[0]) or len(reaptsub[0]):
                        minumrel=99
                        for h in reaptobj[0]:
                            rel=locrel[h,2]
                            if contain(i['boxes'][int(locrel[h,0])],i['boxes'][int(locrel[h,1])],i['boxes'][subinx],w,h):
                                minumrel = minirel(minumrel, rel, df_thrdim)
                        for h in reaptsub[0]:
                            rel=locrel[h,2]
                            if contain(i['boxes'][int(locrel[h,0])],i['boxes'][int(locrel[h,1])],i['boxes'][objinx],w,h):
                                minumrel = minirel(minumrel, rel, df_thrdim)
                        if (minumrel!=99):
                            locall=np.array([[int(objinx), int(subinx), int(minumrel)]])
#                         objname, subname, relname = thename(objinx, subinx, minumrel,i)
#                         print("reapt:",objinx,objname,minumrel,relname,subinx,subname,locate,locrel[h,5])
#                         draw_image(i['img_path'], i,objname,subname)
                            writedics[inx]['relations']=np.append(writedics[inx]['relations'],locall,axis = 0)
            lastrel=len(locrel)
    with open(writedicsname + '.pkl', 'wb') as f:
        pickle.dump(writedics, f)
```

```python
def thrdim_iolcx(pk_dir, thrdimname, writedicsname): #添加新标签，针对overlap
    writedics, df_thrdim = readpdc(pk_dir, thrdimname)
    for inx, i in tqdm(enumerate(writedics[:])): #读取字典每一个图片数据
        w=i['width']
        h=i['height']
        locrel=[] #先保存真实标注
        relations=i['relations']
        for j_inx, j in enumerate(relations): #标注关系对的位置关系
            objinx,subinx,rel = j
            obj=i['labels'][objinx]
            sub=i['labels'][subinx]
            locate=calloc(i['boxes'][objinx],i['boxes'][subinx],w,h)
            locrel.append([objinx,subinx,rel,obj,sub,locate])
        locrel=np.array(locrel)
        for k_inx, k in enumerate(locrel): #枚举标注关系对
            objinx,subinx,thisrel,obj,sub,locate=k
            if (locate==0):
                reaptobj=np.where( (locrel[:,3]==obj) & (locrel[:,4]==sub) & (locrel[:,5]==locate) & (locrel[:,0]==objinx) ) #obj不变，sub变
                reaptsub=np.where( (locrel[:,3]==obj) & (locrel[:,4]==sub) & (locrel[:,5]==locate) & (locrel[:,1]==subinx) ) #sub不变，obj变  
            elif (locate>0):
                reaptobj=np.where( (locrel[:,3]==obj) & (locrel[:,4]==sub) & (locrel[:,5]>0) & \
                                      (locrel[:,5]<=locate) & (locrel[:,0]==objinx) ) #obj不变，sub变
                reaptsub=np.where( (locrel[:,3]==obj) & (locrel[:,4]==sub) & (locrel[:,5]>0) & \
                                      (locrel[:,5]<=locate) & (locrel[:,1]==subinx) ) #sub不变，obj变
            else:
                reaptobj=np.where( (locrel[:,3]==obj) & (locrel[:,4]==sub) & (locrel[:,5]<0) & \
                                      (locrel[:,5]>=locate) & (locrel[:,0]==objinx) ) #obj不变，sub变
                reaptsub=np.where( (locrel[:,3]==obj) & (locrel[:,4]==sub) & (locrel[:,5]<0) & \
                                      (locrel[:,5]>=locate) & (locrel[:,1]==subinx) ) #sub不变，obj变
            if len(reaptobj[0]) or len(reaptsub[0]):
                    minumrel=thisrel
                    for h in reaptobj[0]:
                        rel=locrel[h,2]
                        if contain(i['boxes'][int(locrel[h,0])],i['boxes'][int(locrel[h,1])],i['boxes'][int(subinx)],w,h):
                            minumrel = minirel(minumrel, rel, df_thrdim)
                    for h in reaptsub[0]:
                        rel=locrel[h,2]
                        if contain(i['boxes'][int(locrel[h,0])],i['boxes'][int(locrel[h,1])],i['boxes'][int(objinx)],w,h):
                            minumrel = minirel(minumrel, rel, df_thrdim)
                    if (minumrel!=thisrel):
#                         objname, subname, relname = thename(int(objinx), int(subinx), int(minumrel),i)
#                         print("reapt:",objinx,objname,thisrel,minumrel,relname,subinx,subname)
#                         draw_image(i['img_path'], i,objname,subname)
                        writedics[inx]['relations'][k_inx][2]=minumrel
    with open(writedicsname + '.pkl', 'wb') as f:
        pickle.dump(writedics, f)
```
